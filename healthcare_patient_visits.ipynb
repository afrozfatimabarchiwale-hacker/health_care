```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt

#load dataset
df=pd.read_csv("healthcare_patient_visits.csv")
print(df.to_string)

#first 5 rows
print(df.head)
#data info
print(df.info())

#summary Statistics
print(df.describe())

#Handle Missing Values
print(df.isnull().sum())

#Replace missing marks with colum mean
df.fillna(df.mean(numeric_only=True),inplace=True)
print(df)

#Average Treatmentcost by condition
PatientID=["TreatmentCost","City","SystolicBP","Condition"]

print("\n ---Average TreatmentCst by Condition")
print(df["PatientID"].mean())

#Visualization
#Bar Chart: TreatmentCost vs Condition
plt.figure(figsize=(12,5))
plt.bar(df["TreatmentCost"],df["Condition"])
plt.xlabel("TreatmentCost")
plt.ylabel("Condition")
plt.title("TreatmentCost by Condition")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

#histogram chart: Age distribution
plt.figure(figsize=(12,5))
plt.hist(df["Age"],edgecolor='black')
plt.xlabel("Age")
plt.ylabel("Number of Patients")
plt.title("Age Distribution")
plt.tight_layout()
plt.show()

# Boxplot: TreatmentCost by Smoking Status
plt.figure(figsize=(10,6))
df.boxplot(column='TreatmentCost', by='SmokingStatus')
plt.xlabel("Smoking Status")
plt.ylabel("Treatment Cost")
plt.title("Treatment Cost by Smoking Status")  
plt.tight_layout()
plt.show()

```
